```python
import numpy as np

W = 4
H = 4

Q = np.zeros((W*2, H), dtype=float)
K = np.zeros((W*2, H), dtype=float)
V = np.zeros((W*2, H), dtype=float)

for i in range(W*2):
    for j in range(H):
        Q[i, j] = i*H + j + 1
        K[i, j] = i*H + j + 2
        V[i, j] = i*H + j

print("Q:\n", Q)
print("K:\n", K)
print("V:\n", V)
```

```python
import math

def Attention_Core(Q, K, V, scale=3):
    diff=abs(np.sum(Q-K))
    S=math.e**(-scale*diff)

    print("S: ",S)

    Z=S*V
    print("Z: ",Z)
    print("--------------------------------------")

    return np.array(S,dtype=float),np.array(Z,dtype=float)
```

```python
S = np.zeros((W*2), dtype=float)
Z = np.zeros((W*2, H), dtype=float)

for i in range(W*2):
    S[i],Z[i]=Attention_Core(Q[i], K[i], V[i])
```

```python
Zsum=np.zeros((H), dtype=float)

for i in range(W*2):
    Zsum+=Z[i]

print("Zsum: ",Zsum)
```
